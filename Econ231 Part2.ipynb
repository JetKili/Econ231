# Econ231 Portfolio Frontier

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\Moses\Desktop\econ231_2020spring\PS4\ps4.xls',sheet_name=1)
```

```python
df.set_index('Date',inplace=True)
```

```python
df
```

```python
df['r_CXSE'] = df.CXSE/df.CXSE.shift(-1)-1
df['r_SPTL'] = df.SPTL/df.SPTL.shift(-1)-1
```

```python
df
```

```python
df.describe()
```

```python
mean_r_CXSE=df['r_CXSE'].mean()
mean_r_CXSE
```

```python
mean_r_SPTL=df['r_SPTL'].mean()
mean_r_SPTL
```

```python
var_r_CXSE=df['r_CXSE'].var()
var_r_CXSE
```

```python
var_r_SPTL=df['r_SPTL'].var()
var_r_SPTL
```

```python
cov_r=df[['r_CXSE','r_SPTL']].cov().iloc[1,0]
cov_r
```

```python
d={'weight_CXSE':np.arange(0,1.05,0.05),'weight_SPTL':1-np.arange(0,1.05,0.05)}
protfolio=pd.DataFrame(data=d)
protfolio
```

```python
protfolio['return']=protfolio.weight_CXSE*mean_r_CXSE+protfolio.weight_SPTL*mean_r_SPTL
protfolio['std_dev']=(protfolio.weight_CXSE**2*var_r_CXSE+protfolio.weight_SPTL**2*var_r_SPTL+2*protfolio.weight_CXSE*protfolio.weight_SPTL*cov_r)**0.5
protfolio
```

```python
from scipy.optimize import minimize
neg_sharpe_ratio = lambda weight: -(weight*mean_r_CXSE+(1-weight)*mean_r_SPTL-0.005)/(weight**2*var_r_CXSE+(1-weight)**2*var_r_SPTL+2*weight*(1-weight)*cov_r)**0.5
res=minimize(neg_sharpe_ratio,0.5)
sharpe_ratio=-res.fun
optimal_weight=res.x[0]
res
```

```python
f={'std_dev':np.arange(0,0.07,0.01)}
frontier=pd.DataFrame(data=f)
frontier['return']=0.005+sharpe_ratio*frontier.std_dev
frontier
```

```python
opt_return=optimal_weight*mean_r_CXSE+(1-optimal_weight)*mean_r_SPTL
opt_risk=(optimal_weight**2*var_r_CXSE+(1-optimal_weight)**2*var_r_SPTL+2*optimal_weight*(1-optimal_weight)*cov_r)**0.5
(opt_risk,opt_return)
```

```python
plt.plot(protfolio['std_dev'],protfolio['return'],'bo-', frontier['std_dev'],frontier['return'],'y-',0,0.005,'go',opt_risk,opt_return,'ro')
plt.annotate('risk free rate', (0.003, 0.005))
plt.annotate('optimal sharpe ratio', (opt_risk+0.004,opt_return-0.0001))
plt.title('Portfolio Frontier')
plt.xlabel('Risk')
plt.ylabel('Return')
plt.show()
```
